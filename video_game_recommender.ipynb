## Recommender Systems
In general, recommender systems are machine learning systems that have algorithms used to suggest relevant items to users such as, recommended movies to watch, products to buy and books to read (Mishra, 2021). Companies like Netflix and Amazon use recommender systems to help their users identify the correct product or movies for them. It also plays an important role in helping users have a better user experience.

The recommender system handles the large volume of information presented to it by filtering the most important information based on data provided by the user and many other factors. It figures out the match between the user and product or service and assigns the similarities between users and product or service for recommendation.


### Recommender Systems - Content-based Approach
A content-based recommender system tries to recommend items to users based on their profile. The users profile revolves around that users’ preferences. It is shaped based on user ratings including the number of times the user has clicked on different items or even liked those items (Mishra, 2021). The recommendation process is based on the similarity between those items. Similarity of items is measured based on the similarity in the content of those items. For example, a list of movie’s contents would be its genre, tags, and directors. If there were a list of four movies and the user has rated the first two movies, if the third movie was like the first two movies in terms of genre, the engine will also recommend the third movie to the user. In general, this is what a content-based recommender system does. While the user provides more data where if it is giving a rating or clicking a link, the engine becomes more accurate (Mishra, 2021). Figure 1 below shows an accurate image representation of a content-based approach recommender system.



The content-based recommendation system works on two methods, both use different models and algorithms. One uses the vector spacing method and the other uses a classifications model. For example, using the vector spacing method, a ranking system where a user reviews movies on a scale of 0 to 10. A user vector is created which ranks the information provided by them. After, an item vector is created where movies are ranked according to their genres. With the vector, every movie name is assigned a value by multiplying and getting the dot product of the user and item vector. This value is then used for recommendations. The dot products of all available movies searched by the user are ranked and according to it the top 10 movies are then assigned (Mishra, 2021).

The classification method used by content-based recommender systems is for example, a movie is considered called The Conjuring. Checking that the director’s name is not James Wan. Then, the genre is not horror, nor it is a movie the user reviewed. With these factors it is concluded that the movie should not be recommended to the user.


### Recommender Systems - Collaborative Approach
Collaborative approach to recommender systems is that relationships exist between products and people’s interests (Mishra, 2021). Many recommendation systems use collaborative filtering to find these relationships and to give an accurate recommendation of a product that the user might be interested in. There are two approaches to collaborative filtering, one is user-based and the other is item-based. User-based collaborative filtering is based on the user similarity. Item-based collaborative filtering is based on the similarity among items (Mishra, 2021). Figure 2 below shows an accurate image representation of a collaborative approach recommender system.

In user-based collaborative filtering, there is an active user for whom the recommendation will be aimed towards. The collaborative filtering system will look for users who are similar. Using the list of movies example, it would be users who share the active user’s rating patterns. This filtering basis the similarity on things like preference and choices users make when buying or watching something. For example, movies that similar users have rated highly. It then uses the ratings from the similar users to predict the possible ratings by the active user for a movie they have not watched before.

Item-based collaborative filtering similar items build areas on the behaviour of users (Mishra, 2021). This is not based on their contents but based on the preferences of users, for example, movie one and movie three are considered ‘neighbours’ as they were both positively rated by user one and user two meaning user three can be recommended movie one as that user has already shown interest in movie three.

### Video Game Recommender System
The recommender system developed for this assignment is a content-based recommender system. It recommends an item based on its features and how similar they are to features of other items in the dataset using the nearest neighbour algorithm (K-Nearest Neighbor(KNN) Javatpoint, 2021). This recommender system will recommend video games based off its sales around the world such as in the EU, North America, and Japan.

### Nearest Neighbor Algorithm (KNN)
- Is also known as a memory-based system as it memorizes instances and then recommends an item (a single instance) based on how quantitatively similar it is to a new instance.
- KNN is a non-parametric algorithm, meaning it does not make any assumption on underlying data.
- Sometimes called a lazy learner algorithm because it does not learn from the training set immediately instead it stores the dataset and at the time of classification, it then performs an action on the dataset.
- The KNN algorithm is simple to implement however computation cost is high because of calculating the distance between the data points for all the training samples.

### Video Game Recommender System - Background Story
A games collector is asking what games the seller has in their inventory depending on their sales around the world. Their specifications are games sold in the EU of 10 million sales, 15 million sales in North America, and 5 million sales in Japan. They would like to purchase the game nearest to these specifications and hope to add it to their collection.

```python
#Importing libraries for the analysis 

#Numerical Python as np and pandas as pd
#Importing libraries/packages

#Numpy is a library consisting of mutli-dimensional array objects and has routines for processing those arrays
#Used to perform mathematical and logical operations on arrays
import numpy as np

#Python toolkit used for data manipulation and analysis
import pandas as pd

#Importing scikit-learn and nearest neighbors
import sklearn
from sklearn.neighbors import NearestNeighbors
```

```python
#Read the vgsales dataset
#Read csv function
games = pd.read_csv('~/AI_Assignment_1/videogame-metadata/vgsales.csv')

#Naming columns
games.columns = ['Rank', 'Name', 'Platform', 'Year', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']

#Head method to display columns
#Returns the first n rows (5 by default)
games.head()
```

```python
#Test point for shoppers specifications for games sales

#Shopper is looking for with sales in the EU of 10 million,
#North America sales of 15 million
#Japan sales of 5 million
t = [10, 15, 5]

#Define dataset
#Special indexer to select only variables we need
#.loc accesses a group of rows and columns by label
X = games.loc[:,('EU_Sales', 'NA_Sales', 'JP_Sales')].values

#Look at the first 5 records
X[0:5]

#Subset is created in Output 3 from the original cars dataframe (see Output 2)
```

```python
#Building Nearest Neighbor Model

#Instantiating nearest neighbor object (nbrs) by calling nearest neighbor function
#Tells the algorithm to search the dataset and find a single point p that is nearest to the test point t
#.fit fits the model to the data, passing in our X dataset
nbrs = NearestNeighbors(n_neighbors = 1).fit(X)
```

```python
#Calling k neighbors function on the test point t
#Accessing this function from within the nearest neighbor model
#Print result
print(nbrs.kneighbors([t]))

#Returns an array that represents length to point p from test point t 
#Array that contains index value of the nearest point, the record thats index position is 3
```
